```python
# Análisis de variables numéricas
exitosos = df['sol1']
print(exitosos.describe())

def c_skew(x):
    x_mean = np.mean(x.dropna())
    x_std = np.std(x.dropna())
    m3 = np.mean((x.dropna() - x_mean) ** 3)
    skew = m3 / (x_std ** 3)
    return skew

skewness = c_skew(df['sol1'])
print('sesgo: ',skewness)
print("sesgo en porcentaje: {:.2f}%".format(skewness))
```

```python
# Coeficiente de asimetría
def c_skew(x):
    m3 = np.mean((x - np.mean(x, axis=0)) ** 3, axis=0)
    skew = m3 / (np.std(x, axis=0) ** 3)
    skew_percent = skew * 100
    return skew, skew_percent
skewness, skewness_percent = c_skew(df['sol1'])
print("Coeficiente de asimetría:", skewness)
print("Coeficiente de asimetría en porcentaje: {:.2f}%".format(skewness_percent))
```

```python
# Coeficiente de variación
def coef_var(x):
    std = np.std(x, axis=0)
    mean = np.mean(x, axis=0)
    cv = std / mean
    cv_percent = cv * 100
    return cv, cv_percent

cv, cv_percent = coef_var(df['sol1'])
print("Coeficiente de variación:", cv)
print("Coeficiente de variación en porcentaje: {:.2f}%".format(cv_percent))
```

```python
#Obteniendo amplitud
n = len(data2['sol1'])
amplitud = (np.nanmax(data2['sol1']) - np.nanmin(data2['sol1'])) / (1 + 3.322 * np.log10(n))
print('Amplitud: ', amplitud)
print("Amplitud en porcentaje: {:.2f}%".format(amplitud))
```

```python
# Tabla de frecuencias
data2['sol1.cut'] = pd.cut(data2['sol1'], bins=np.arange(0, 7, 0.56))
f_i = data2['sol1.cut'].value_counts().sort_index().values
F_i = f_i.cumsum()
h_i = f_i / len(data2['sol1'])
H_i = F_i / len(data2['sol1'])

table = pd.DataFrame({'f_i': f_i, 'F_i': F_i, 'h_i': h_i, 'H_i': H_i}, index=data2['sol1.cut'].value_counts().sort_index().index)
display(table)
```

```python
mediana = np.median(data2['sol1'])
maximo = np.max(data2['sol1'])
intervalo_mediana = table.loc[table.index[np.where(np.cumsum(f_i) / np.sum(f_i) >= 0.5)[0][0]]]
intervalo_maximo = table.loc[table.index[-1]]
print("Mediana:", mediana)
print("Intervalo de la mediana:", intervalo_mediana)
print("Máximo:", maximo)
print("Intervalo del máximo:", intervalo_maximo)
```

```python
# Calcular estadísticos descriptivos
stats = df.describe()

# Calcular límites para detectar valores atípicos (ejemplo: utilizando el rango intercuartil)
Q1 = stats.loc['25%']
Q3 = stats.loc['75%']
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identificar valores atípicos
outliers = df[(df < lower_bound) | (df > upper_bound)]
display(outliers)
```

```python
# Filtrar los datos para obtener solo los aprobados (aprobado = 1)
aprobados = df[df['aprobado'] == 1]

# Filtrar los datos para obtener solo los reprobados (aprobado = 0)
reprobados = df[df['aprobado'] == 0]

# Contar el número de aprobados y reprobados
num_aprobados = len(aprobados)
num_reprobados = len(reprobados)

# Crear una lista con los datos para el gráfico de torta
datos_torta = [num_aprobados, num_reprobados]

# Etiquetas para el gráfico de torta
etiquetas = ['Aprobados', 'Reprobados']

# Colores para cada porción del gráfico de torta
colores = ['#33FF57', '#FF5733']

# Crear el gráfico de torta
plt.figure(figsize=(8, 8))
plt.pie(datos_torta, labels=etiquetas, colors=colores, autopct='%1.0f%%', startangle=90)

# Agregar título al gráfico
plt.title('Porcentaje de Aprobados y Reprobados')

# Mostrar el gráfico
plt.show()
```

```python
# Calcular el total de éxitos y fallos
total_exitosos = df['exitosos'].sum()
total_fallidos = df['fallidos'].sum()

# Calcular la proporción de aprobados y reprobados
total_aprobados = df['aprobado'].sum()
total_reprobados = len(df) - total_aprobados

# Calcular las proporciones para el gráfico de "bar of pie"
overall_ratios = [total_exitosos / (total_exitosos + total_fallidos),
                  total_fallidos / (total_exitosos + total_fallidos)]
labels = ['Exitosos', 'Fallidos']
explode = [0.1, 0]  # Explode el primer sector del gráfico

# Crear el gráfico de "bar of pie"
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(9, 5))
fig.subplots_adjust(wspace=0)

# Gráfico circular
angle = -180 * overall_ratios[0]
wedges, *_ = ax1.pie(overall_ratios, autopct='%1.0f%%', startangle=angle, labels=labels, explode=explode)

# Gráfico de barras
aprobados_ratios = [total_aprobados / len(df), total_reprobados / len(df)]
aprobados_labels = ['Aprobados', 'Reprobados']
bottom = 1
width = 0.2

for j, (height, label) in enumerate(reversed([*zip(aprobados_ratios, aprobados_labels)])):
    bottom -= height
    bc = ax2.bar(0, height, width, bottom=bottom, color='C0', label=label, alpha=0.1 + 0.25 * j)
    ax2.bar_label(bc, labels=[f"{height:.0%}"], label_type='center')

ax2.set_title('Aprobados y Reprobados')
ax2.legend()
ax2.axis('off')
ax2.set_xlim(-2.5 * width, 2.5 * width)

# Conexiones entre los gráficos
theta1, theta2 = wedges[0].theta1, wedges[0].theta2
center, r = wedges[0].center, wedges[0].r
bar_height = sum(aprobados_ratios)

x = r * np.cos(np.pi / 180 * theta2) + center[0]
y = r * np.sin(np.pi / 180 * theta2) + center[1]
con = ConnectionPatch(xyA=(-width / 2, bar_height), coordsA=ax2.transData,
                      xyB=(x, y), coordsB=ax1.transData)
con.set_color([0, 0, 0])
con.set_linewidth(4)
ax2.add_artist(con)

x = r * np.cos(np.pi / 180 * theta1) + center[0]
y = r * np.sin(np.pi / 180 * theta1) + center[1]
con = ConnectionPatch(xyA=(-width / 2, 0), coordsA=ax2.transData,
                      xyB=(x, y), coordsB=ax1.transData)
con.set_color([0, 0, 0])
ax2.add_artist(con)
con.set_linewidth(4)

plt.show()
```
